```python
import numpy as np
import cv2
```

```python
img = cv2.imread('images/wall street.jpg')
```

```python
# Show WS photo

cv2.imshow('Wall Street', img)
cv2.waitKey(10_000)
cv2.destroyWindow(winname='Wall Street')
```

```python
# Show Black box (BB)

photo = np.zeros(shape=(450, 450, 3), dtype='uint8')

cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```

```python
# Draw square in BB

photo = np.zeros(shape=(450, 450, 3), dtype='uint8')
cv2.rectangle(img=photo, 
              pt1=(100, 100), 
              pt2=(300, 300), 
              color=(0, 255, 0), 
              thickness=3)
cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```

```python
# Paint the entire square

photo = np.zeros(shape=(450, 450, 3), dtype='uint8')
cv2.rectangle(img=photo, 
              pt1=(100, 100), 
              pt2=(300, 300), 
              color=(0, 255, 0), 
              thickness=cv2.FILLED)  # Use attribute .FILLED
cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```

```python
# Display line in the middle of square

photo = np.zeros(shape=(450, 450, 3), dtype='uint8')
cv2.line(img=photo, 
         pt1=(0, photo.shape[0] // 2), 
         pt2=(photo.shape[0], photo.shape[0] // 2), 
         color=(0, 0, 255), 
         thickness=5)
cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```

```python
# Display filled circle in the middle of black box

photo = np.zeros(shape=(450, 450, 3), dtype='uint8')
cv2.circle(img=photo, 
           center=(photo.shape[1] // 2, photo.shape[0] // 2),
           radius=100,
           color=(0, 165, 225),
           thickness=cv2.FILLED)
cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```

```python
cv2.putText(img=photo, 
            text='Alexander Kapturov',
            org=(100, 100),
            fontFace=cv2.FONT_HERSHEY_TRIPLEX,
            fontScale=1,
            color=(255, 0, 0),
            thickness=2)
cv2.imshow('Black box', photo)
cv2.waitKey(5_000)
cv2.destroyWindow(winname='Black box')
```
