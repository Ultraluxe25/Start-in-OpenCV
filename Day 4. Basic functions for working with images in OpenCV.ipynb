````python
import cv2
````

````python
img = cv2.imread('images/wall street.jpg')
````

````python
cv2.imshow('Wall Street', img)
cv2.waitKey(5_000)
````

````python
img.shape  # (height, width, 3 is RGB)
````

````python
# Resize picture

new_img = cv2.resize(img, (450, 300))
cv2.imshow('Small WS', new_img)
cv2.waitKey(5_000)
````

````python
# Resize correctry

new_img_2 = cv2.resize(img, (img.shape[1] // 2, img.shape[0] // 2))
cv2.imshow('Big WS', new_img_2)
cv2.waitKey(5_000)
````

````python
# Image cropping

crop_img = cv2.resize(img, (600, 900))
cv2.imshow('Big WS', crop_img[200:400, 300:600])
cv2.waitKey(5_000)
````

````python
# Image blur
blur = cv2.GaussianBlur(img, (9, 9), 2)
cv2.imshow('blur', blur)
cv2.waitKey(5_000)
````

# [cv2 GaussianBlur()](https://www.tutorialkart.com/opencv/python/opencv-python-gaussian-image-smoothing/#gsc.tab=0)

````python
# Black and white format (processing becoming easier)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # There is only one gray tier (not 3 like in RGB)
cv2.imshow('Gray WS', gray)
cv2.waitKey(5_000)
````

# [OpenCV | cv2.cvtColor()](https://www.geeksforgeeks.org/python-opencv-cv2-cvtcolor-method/)

````python
angle = cv2.Canny(img, 90, 90)  # Image become binary colored (black and white)
cv2.imshow('WS', angle)
cv2.waitKey(5_000)
````

````python
# Create matrix for experiment
import numpy as np

matrix = np.ones((5, 5), dtype=np.uint8)
matrix
````

**np.uint8** is one of the predefined data types in NumPy that provides fast and efficient work with images, pixels, and other data that can be represented in the range 0 to 255.

````python
# Dilates an image
grey_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grey_angle = cv2.Canny(grey_image, 300, 300)
image = cv2.dilate(grey_angle, kernel=matrix, iterations=1)
cv2.imshow('dilate', image)
cv2.waitKey(5_000)
cv2.destroyAllWindows()
````

# [Cv2 Dilate](https://dev-gang.ru/article/python-cv-dilate-rasshirenie-izobrazhenii-s-ispolzovaniem-opencv-sp8sujfqf9/)

````python
picture = cv2.erode(src=img, kernel=matrix, iterations=3)
cv2.imshow('WS last', picture)
cv2.waitKey(5_000)
cv2.destroyWindow('WS last')
````
