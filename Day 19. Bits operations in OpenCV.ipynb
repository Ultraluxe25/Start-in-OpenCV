```python
import numpy as np
import cv2
```

```python
img = np.zeros(shape=(350, 350), dtype='uint8')
img
```

```python
circle = cv2.circle(img.copy(), center=(0, 0), radius=80, color=255, thickness=-1)
square = cv2.rectangle(img.copy(), pt1=(25, 25), pt2=(250, 250), color=255, thickness=-1)
```

```python
# Circle and square
cv2.imshow('Circle', circle)
cv2.waitKey(5_000)

cv2.imshow(winname='Square', mat=square)
cv2.waitKey(5_000)
cv2.destroyAllWindows()
```

```python
# Bitwise AND, OR, XOR, NOT
img = cv2.bitwise_and(src1=circle, src2=square)
cv2.imshow(winname='AND', mat=img)
cv2.waitKey(5_000)

img = cv2.bitwise_or(src1=circle, src2=square)
cv2.imshow(winname='OR', mat=img)
cv2.waitKey(5_000)

img = cv2.bitwise_xor(src1=circle, src2=square)
cv2.imshow(winname='XOR', mat=img)
cv2.waitKey(5_000)

img = cv2.bitwise_not(src=square)
cv2.imshow(winname='NOT', mat=img)
cv2.waitKey(5_000)
cv2.destroyAllWindows()
```

```python
photo = cv2.imread('images/Lesson 1.png')
img = np.zeros(shape=photo.shape[:2], dtype='uint8')
circle = cv2.circle(img.copy(), center=(200, 300), radius=120, color=255, thickness=-1)

img = cv2.bitwise_and(src1=photo, src2=photo, mask=circle)
cv2.imshow(winname='photo AND', mat=img)
cv2.waitKey(5_000)
cv2.destroyAllWindows()
```

 # Fullbody detection

```python
# My robot team!
img = cv2.imread('images/My robot team.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

bodies = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
results = bodies.detectMultiScale(gray, scaleFactor=1.06, minNeighbors=6)

for (x, y, w, h) in results:
    cv2.rectangle(img, pt1=(x, y), pt2=(x + w, y + h), color=(0, 0, 255), thickness=2)

cv2.imshow('Team', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

# Code below from this [source](https://www.tutorialspoint.com/how-to-detect-humans-in-an-image-in-opencv-python)

```python
# import required libraries
import cv2

# Reading the Image
image = cv2.imread('images/friends.jpg')
image = cv2.resize(image, (image.shape[1] // 2, image.shape[0] // 2))

# initialize the HOG descriptor
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# detect humans in input image
(humans, _) = hog.detectMultiScale(image, winStride=(4, 4),
padding=(8, 8), scale=1.2)

# getting no. of human detected
print('Human Detected : ', len(humans))

# loop over all detected humans
for (x, y, w, h) in humans:
   pad_w, pad_h = int(0.15 * w), int(0.01 * h)
   cv2.rectangle(image, (x + pad_w, y + pad_h), (x + w - pad_w, y + h - pad_h), (0, 255, 0), 2)

# display the output image
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
