```python
import numpy as np
import cv2
```

```python
# HSV format
img = cv2.imread('images/wall street.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('Wall Street', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

```python
# LAB format
img = cv2.imread('images/wall street.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imshow('Wall Street', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

```python
# Return to initial format (BGR)
img = cv2.cvtColor(img, cv2.COLOR_LAB2BGR)
cv2.imshow('Wall Street', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

```python
# If you will use another library like matplotlib or seaborn then convert to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow('Wall Street', img)
cv2.waitKey(10_000)
cv2.destroyAllWindows()
```

```python
# Display different layers of image
img = cv2.imread('images/wall street.jpg')
blue, green, red = cv2.split(img)

cv2.imshow('Wall Street', blue)
cv2.waitKey(3_000)
cv2.destroyAllWindows()

cv2.imshow('Wall Street', green)
cv2.waitKey(3_000)
cv2.destroyAllWindows()

cv2.imshow('Wall Street', red)
cv2.waitKey(3_000)
cv2.destroyAllWindows()
```

Colors that are not far away to red (to green, to blue) convert into white, other convert into shades of gray or black

```python
# Display different types of layer's merges
merge_1 = cv2.merge([red, green, blue])
cv2.imshow('Wall Street', merge_1)
cv2.waitKey(5_000)
cv2.destroyAllWindows()

merge_2 = cv2.merge([red, blue, green])
cv2.imshow('Wall Street', merge_2)
cv2.waitKey(5_000)
cv2.destroyAllWindows()

merge_3 = cv2.merge([blue, green, red])
cv2.imshow('Wall Street', merge_3)
cv2.waitKey(5_000)
cv2.destroyAllWindows()

merge_4 = cv2.merge([blue, red, green])
cv2.imshow('Wall Street', merge_4)
cv2.waitKey(5_000)
cv2.destroyAllWindows()

merge_5 = cv2.merge([green, blue, red])
cv2.imshow('Wall Street', merge_5)
cv2.waitKey(5_000)
cv2.destroyAllWindows()

merge_6 = cv2.merge([green, red, blue])
cv2.imshow('Wall Street', merge_6)
cv2.waitKey(5_000)
cv2.destroyAllWindows()
```
